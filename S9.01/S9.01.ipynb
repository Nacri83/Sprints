2.  Convertidor de temperatura 
Hay varias unidades de temperatura utilizadas en diferentes contextos y regiones. Los m√°s comunes son Celsius (¬∞C), Fahrenheit (¬∞F) y Kelvin (K). Tambi√©n hay otras unidades como Rankine (¬∞Ra) y R√©aumur (¬∞Re). 

Seleccione al menos 2 convertidores, de modo que al entrar una temperatura devuelva al menos dos conversiones, de modo que se puedan guardar (recuerde que una impresi√≥n () nunca se puede guardar). 

Sugerencia: Intenta validar los datos con antelaci√≥n, para enviar un mensaje de advertencia si los datos introducidos por el usuario no est√°n en el formato adecuado.

 (EXTRA): Piense en una manera de almacenar todas las conversiones posibles en un solo objeto (Lista? ¬øUn diccionario? DataFrame?) En lugar de escribir muchos si dependen de la temperatura de origen y la temperatura de destino.

```python
# Definici√≥n de la funci√≥n
def conversor_temperatura(valor, unidad):
# Validaci√≥n de datos
    if unidad not in ["C","F","K","Ra","Re"]:
        print('Error: no es una unidad valida de temperatura')
        return
 # Una instancia es un (valor) concreto que pertenece a un (tipo de dato).
    if not isinstance(valor,(int,float)):          
        print('Esrror: no es un valor n√∫merico')
        return
 # Condicional multiple para decidir formula a usar   
    if unidad == "C":
        celsius = valor
    elif unidad == "F":
        celsius = (valor - 32) * 5 / 9
    elif unidad == "K":
        celsius = valor - 273.15
    elif unidad == "RA":
        celsius = (valor - 491.67) * 5 / 9
    elif unidad == "RE":
        celsius = valor * 5 / 4


    # Diccionario de conversiones en funci√≥n de celsius

    conversiones = {
        "C": celsius,
        "F": (celsius * 9 / 5) + 32,
        "K": celsius + 273.15,
        "RA": (celsius + 273.15) * 9 / 5,
        "RE": celsius * 4 / 5
    }

    return conversiones
#-----------------------------------------------------------------------------------------

# EJECUCI√ìN DEL PROGRAMA


valor = float(input("Introduce la temperatura: "))
unidad = input("Introduce la unidad (C, F, K, Ra, Re): ").upper()

resultado = conversor_temperatura(valor, unidad)

if resultado is not None:
    print("Conversiones:")
    for u, v in resultado.items():
        print(f"{u}: {v}")
```

```python
import string # Si no importo string debo guardar en una variable toda una cadena de signos de puntuaci√≥n

def contador_palabras(texto):
    
     # VALIDACI√ìN

    if not isinstance(texto, str):
        print("Error: la entrada debe ser un texto.")
        return
    
    if texto.strip() == "":
        print("Error: el texto est√° vac√≠o.")
        return

    # NORMALIZACI√ìN
  
    texto = texto.lower()

    # LIMPIEZA DE SIGNOS
 
    for signo in string.punctuation:
        texto = texto.replace(signo, "")


    # SEPARAR EN PALABRAS

    palabras = texto.split()

    # CONTADOR DE PALABRAS
  
    contador = {}

    for palabra in palabras:
        if palabra in contador:
            contador[palabra] += 1
        else:
            contador[palabra] = 1

    # EXTRA: LONGITUD PROMEDIO

    total_letras = 0
    for palabra in palabras:
        total_letras += len(palabra)

    longitud_promedio = total_letras / len(palabras)

    return contador, longitud_promedio
```

```python
import warnings


def diccionario_inverso(dic):                                  # L√ìGICA: FUNCI√ìN
    invertido = {}                                             # L√ìGICA: DICCIONARIO
    valores_duplicados = []                                    # L√ìGICA: LISTA

    for clave, valor in dic.items():                           # L√ìGICA: BUCLE
        
        if valor not in invertido:                             # L√ìGICA: CONDICIONAL
            invertido[valor] = clave

        else:
            # Si el valor ya existe, control de duplicados
            if not isinstance(invertido[valor], list):         # L√ìGICA: VALIDACI√ìN DE TIPO
                valores_duplicados.append(valor)
                invertido[valor] = [invertido[valor]]
            
            invertido[valor].append(clave)                     # L√ìGICA: AGREGACI√ìN A LISTA

    # Aviso usando la librer√≠a warnings (sin bandera)          #LOGICA DE COMUNICACI√ìN ESTRUCTURADA
    if valores_duplicados:
        warnings.warn(
            f"El diccionario original contiene valores duplicados: {valores_duplicados}",
            UserWarning
        )

    return invertido                                           # L√ìGICA: RETORNO


# ===== PROGRAMA DE EJECUCI√ìN =====
print(diccionario_inverso({'a': 1, 'b': 2, 'c': 3}))
# {1: 'a', 2: 'b', 3: 'c'}

print(diccionario_inverso({'x': 'apple', 'y': 'banana', 'z': 'banana'}))
# {'apple': 'x', 'banana': ['y', 'z']}
```

2. Conversi√≥n de tipos de datos

El cliente recibe una lista de datos y necesita generar dos listas:

A)la primera, donde estar√°n todos los elementos que se pudieron convertir a flotantes (float)

B)la segunda, donde estar√°n los elementos que no se pudieron convertir

Ejemplo de la lista que recibe el cliente:

['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 4, 'not-a-number'), [1, 2, '3', '3.4']]


```python
def conversion(lista_datos):                                               #LOGICA FUNCI√ìN

    convertidos = []
    no_convertidos = []

    for elemento in lista_datos:                                           #LOGICA DE BUCLE

        # Si el elemento es una lista o tupla, recorrerlo por dentro
        if isinstance(elemento, (list, tuple)):                           #LOGICA DE VALIDACI√ìN DE TIPOS

            for subelemento in elemento:                                  #LOGICA DE BUCLE

                try:                                                      #LOGICA DE MANEJO DE ERRORES
                    convertidos.append(float(subelemento))
                except:
                    no_convertidos.append(subelemento)

        else:
            try:                                                         #LOGICA DE MANEJO DE ERRORES
                convertidos.append(float(elemento))                      #LOGICA CONVERSI√ìN  Y L√ìGICA DE AGREGACI√ìN LISTA
            except:
                no_convertidos.append(elemento)                          #LOGICA DE FILGTRADO Y L√ìGICA DE AGREGACI√ìN LISTA

    return convertidos, no_convertidos                                  #L√ìGICA DE RETORNO

# EJECUCI√ìN DEL PROGRAMA
datos = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 4, 'not-a-number'), [1, 2, '3', '3.4']]

resultado = conversion(datos)

print("Convertidos:", resultado[0])
print("No convertidos:", resultado[1])
```

1. Generador de Contrase√±as

Explora el funcionamiento del m√≥dulo random de la librer√≠a NumPy. üîó Random module in NumPy

En este punto, el cliente ha detectado un problema con las contrase√±as que utilizan sus trabajadores: Asdf1234, fechas de cumplea√±os o similares. Para solucionarlo, nos ha encargado una funci√≥n en Python que genere contrase√±as m√°s seguras.

   a)La funci√≥n debe depender de los siguientes par√°metros:

   b) longitud (int): Longitud de la contrase√±a

   c) mayusculas (bool = True): Si deben aparecer letras may√∫sculas

   d) minusculas (bool = True): Si deben aparecer letras min√∫sculas

   e) numeros (bool = True): Si deben aparecer n√∫meros

   f) signos (bool = False): Si deben aparecer caracteres especiales (,-$? o similares)

As√≠ pues, si ejecutamos la funci√≥n de la siguiente manera:

        crear_contrasenya(10, True, True, True, True)


Deber√≠amos obtener un output (que debemos poder guardar) del estilo:

        Er,5Vn8P$


Aseg√∫rate de que se cumplan todos los criterios y de que las contrase√±as sean realmente aleatorias.

(EXTRA) Explora c√≥mo podr√≠amos hacer que la funci√≥n copie autom√°ticamente la contrase√±a al portapapeles del ordenador (como si la hubi√©ramos seleccionado y pulsado Ctrl + C).

```python
import numpy as np
import string
import pyperclip


#Definir conjunto de caracteres

MAYUSCULAS = string.ascii_uppercase
MINUSCULAS = string.ascii_lowercase
NUMEROS = string.digits
SIGNOS = string.punctuation

# Definir la funci√≥n
def crear_contrase√±a(longitud,
                      mayusculas=True,
                      minusculas=True,
                      numeros=True,
                      signos=False):
    
# Validaci√≥n b√°sica
    if longitud <= 0:
        raise ValueError("La longitud debe ser mayor que 0")

    if not any([mayusculas, minusculas, numeros, signos]):
        raise ValueError("Debe activarse al menos un tipo de car√°cter")
    
# Garantizar cumplimientos de criterios

    caracteres = []
    bolsa = ""

    if mayusculas:
        caracteres.append(np.random.choice(list(MAYUSCULAS)))
        bolsa += MAYUSCULAS

    if minusculas:
        caracteres.append(np.random.choice(list(MINUSCULAS)))
        bolsa += MINUSCULAS

    if numeros:
        caracteres.append(np.random.choice(list(NUMEROS)))
        bolsa += NUMEROS

    if signos:
        caracteres.append(np.random.choice(list(SIGNOS)))
        bolsa += SIGNOS
    
# Completar hasta la longitud deseada

    while len(caracteres) < longitud:

        caracteres.append(np.random.choice(list(bolsa)))

# Mezclar para m√°xima aletoriedad

    np.random.shuffle(caracteres)

#Convertir a string y devolver resultado

    return "".join(caracteres)

    

#EJECUCION DE PROGRAMA

password = crear_contrase√±a(10, True, True, True, True)
print(password)
pyperclip.copy(password)
```

2. Procesamiento de datos simple

Una compa√±era de trabajo nos ha pedido un favor, aprovechando que sabe que estamos aprendiendo a programar. Tiene un hist√≥rico de partidos de f√∫tbol catal√°n en un archivo, donde est√°n almacenados los nombres de los equipos y los resultados. Necesita que procesemos los datos de forma autom√°tica para extraer la informaci√≥n que necesita.

Utiliza el archivo "historic_partits.txt".

Necesita un programa que devuelva:

El n√∫mero total de goles que ha marcado cada equipo.

El nombre del equipo m√°s goleador.

El nombre del equipo m√°s goleado.

La clasificaci√≥n global (cada victoria: 3 puntos, empate: 1 punto, derrota: 0 puntos).

```python
# =========================
# 1. Inicializar estructura
# =========================

equipos = {}

# =========================
# 2. Abrir y leer el archivo
# =========================

with open(r"C:\Ana_CDPY\Especializacion\S9.01\historic partits.txt", "r", encoding="utf-8") as archivo:
    for linea in archivo:

        linea = linea.strip()

        # Saltar l√≠neas vac√≠as
        if not linea:
            continue

        # Separar la l√≠nea en partes por espacios
        partes = linea.split()

        # Buscar la posici√≥n del marcador (ej: 5-3)
        for i, parte in enumerate(partes):
            if "-" in parte:
                indice_resultado = i
                break

        # Equipo local = todo lo anterior al marcador
        equipo_local = " ".join(partes[:indice_resultado])

        # Resultado
        goles_local, goles_visitante = partes[indice_resultado].split("-")
        goles_local = int(goles_local)
        goles_visitante = int(goles_visitante)

        # Equipo visitante = todo lo posterior al marcador
        equipo_visitante = " ".join(partes[indice_resultado + 1:])

        # =========================
        # 3. Inicializar equipos
        # =========================

        for equipo in [equipo_local, equipo_visitante]:
            if equipo not in equipos:
                equipos[equipo] = {
                    "goles_favor": 0,
                    "goles_contra": 0,
                    "ganados": 0,
                    "empatados": 0,
                    "perdidos": 0,
                    "puntos": 0
                }

        # =========================
        # 4. Actualizar goles
        # =========================

        equipos[equipo_local]["goles_favor"] += goles_local
        equipos[equipo_local]["goles_contra"] += goles_visitante

        equipos[equipo_visitante]["goles_favor"] += goles_visitante
        equipos[equipo_visitante]["goles_contra"] += goles_local

        # =========================
        # 5. Resultado del partido
        # =========================

        if goles_local > goles_visitante:
            equipos[equipo_local]["ganados"] += 1
            equipos[equipo_local]["puntos"] += 3
            equipos[equipo_visitante]["perdidos"] += 1

        elif goles_local < goles_visitante:
            equipos[equipo_visitante]["ganados"] += 1
            equipos[equipo_visitante]["puntos"] += 3
            equipos[equipo_local]["perdidos"] += 1

        else:
            equipos[equipo_local]["empatados"] += 1
            equipos[equipo_visitante]["empatados"] += 1
            equipos[equipo_local]["puntos"] += 1
            equipos[equipo_visitante]["puntos"] += 1

# =========================
# 6. Resultados solicitados
# =========================

print("\nGOLES TOTALES POR EQUIPO")
for equipo, datos in equipos.items():
    print(f"{equipo}: {datos['goles_favor']} goles")

equipo_mas_goleador = max(equipos, key=lambda e: equipos[e]["goles_favor"])
equipo_mas_goleado = max(equipos, key=lambda e: equipos[e]["goles_contra"])

print("\nEquipo m√°s goleador:", equipo_mas_goleador)
print("Equipo m√°s goleado:", equipo_mas_goleado)

# =========================
# 7. Clasificaci√≥n global
# =========================

clasificacion = sorted(
    equipos.items(),
    key=lambda item: item[1]["puntos"],
    reverse=True
)

print("\nCLASIFICACI√ìN GLOBAL")
for posicion, (equipo, datos) in enumerate(clasificacion, start=1):
    print(f"{posicion}. {equipo} - {datos['puntos']} puntos")
```

1. Consulta l’API pública JSONPlaceholder utilitzant el mètode GET per obtenir:

- Llista de publicacions (/posts)
- Llista d’usuaris (/users)
- Llista de tasques (/todos)


```python
import requests
import pandas as pd
```

```python
base_url = "https://jsonplaceholder.typicode.com/"
```

```python
posts = requests.get(f"{base_url}/posts")
users = requests.get(f"{base_url}/users")
todos = requests.get(f"{base_url}/todos")
```

2. Cantidad de recursos y código de estado

```python
print("Posts:", len(posts.json()), "Status:", posts.status_code)
print("Users:", len(users.json()), "Status:", users.status_code)
print("Todos:", len(todos.json()), "Status:", todos.status_code)
```

3. Petición a recurso inexistente (404)

```python
error = requests.get(f"{base_url}/posts/9999")
print("Status code error:", error.status_code)
```

4. Petición POST (crear publicación)

```python
nueva_publicacion = {
    "title": "Mi primera publicación API",
    "body": "Este es un mensaje de prueba",
    "userId": 1
}

respuesta_post = requests.post(f"{base_url}/posts", json=nueva_publicacion)

print("JSON respuesta:", respuesta_post.json())
print("Status:", respuesta_post.status_code)
```

5. Petición PATCH (modificar parcialmente)

```python
modificacion = {
    "title": "Título actualizado"
}

respuesta_patch = requests.patch(f"{base_url}/posts/1", json=modificacion)

print("JSON respuesta:", respuesta_patch.json())
print("Status:", respuesta_patch.status_code)
```

6. Petición DELETE

```python
respuesta_delete = requests.delete(f"{base_url}/posts/1")

print("JSON respuesta:", respuesta_delete.json())
print("Status:", respuesta_delete.status_code)
```

1. Buscar un dataset con package_search y ver detalles con package_show

```python

# Buscar datasets relacionados con accidentes de tráfico
url3 = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/package_search?q=accidents"

respuesta = requests.get(url3)

print("Código de estado:", respuesta.status_code)

datos = respuesta.json()

# Mostrar algunos datasets encontrados
for dataset in datos["result"]["results"][:10]:
    print("Título:", dataset["title"])
    print("ID:", dataset["id"])
    print("--------------")
```

2. Obtener detalles del dataset con package_show

```python

url3 = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/package_show?id=accidents-gu-bcn"

respuesta = requests.get(url3)

datos = respuesta.json()

print("Título:", datos["result"]["title"])
print("Número de recursos:", len(datos["result"]["resources"]))

# Mostrar los primeros 3 recursos
for recurso in datos["result"]["resources"][:3]:
    print("Nombre:", recurso["name"])
    print("Formato:", recurso["format"])
    print("Resource ID:", recurso["id"])
    print("-------------")
```

3. Recuperar al menos 100 registros con datastore_search

```python
import requests

url = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/datastore_search"

params = {
    "resource_id": "bd601249-a019-4798-aa3a-f112d4b6845b",
    "limit": 100
}

respuesta = requests.get(url, params=params)

datos = respuesta.json()

records = datos["result"]["records"]

print("Registros:", len(records))
```

4. L¿Convertir en dataframe

```python
df = pd.DataFrame(records)

df.head()
```

5. Guarda archivo como csv

```python
df.to_csv("bicing_bcn.csv", index=False)
```
